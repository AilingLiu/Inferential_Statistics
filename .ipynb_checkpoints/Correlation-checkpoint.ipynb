```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, t
import pingouin
import matplotlib.dates as mdates
```

## Is the rent price in Chicago correlated to USA rent price?

```python
rent_data = pd.read_csv('https://assets.datacamp.com/production/repositories/6125/datasets/07190c97ff82fd6ced781efa760f448830d24ac1/median_rent.csv')
```

```python
date_columns = rent_data.iloc[:, 6:].columns
chicago = rent_data.query("City == 'Chicago'")[date_columns].values.tolist()[0]
dates = pd.to_datetime(pd.to_datetime(date_columns, format='%B %Y').strftime('%Y-%m-%d').values.tolist())
usa = rent_data[date_columns].mean(axis=0).values.tolist()
```

```python
plt.plot(dates, chicago, label='Chicago')
plt.plot(dates, usa, label='USA')

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))   # show only YYYY
ax.xaxis.set_major_locator(mdates.YearLocator()) 

plt.legend(loc='upper right')
plt.show()
```

```python
from scipy.stats import pearsonr
r, p_value = pearsonr(chicago, usa)
print(r)
print(p_value)
```

As Correlation coefficient is approximately at 0.94, the rent prices of Chicago is strongly correlated to that of USA. The P value has the null hypothesis that these two variables are not correlated, and since the p value is much less then 0.05, we reject the null hypothesis, and conclude that the rent prices of Chicago and USA are correlated. It's important to look at the R value directly to understand the relationship direction, as either strong negative or positive values will end up having small P value.

### Drawbacks of correlation

- Correlated data is not independent
- Violates assumptions of some hypothesis tests
- Data can be correlated with itself

---> validate **Autocorrelation**

### Autocorrelation
Do rent prices tend to show a similar pattern when compared year-over-year? If you take the rent prices in Los Angeles and compare them to rent prices one year in the past, will you see a significant relationship? In other words, do rent prices in Los Angeles demonstrate autocorrelation?

```python
la_rents = rent_data.query("City == 'Los Angeles'")[date_columns].values.tolist()[0]
```

```python
# Select all but the first twelve rents
la_rents_initial = la_rents[12:]

# Select all but the last twelve rents (12 month lag)
la_rents_lag = la_rents[:-12]

# Compute the correlation between the initial values and the lagged values
r, p_value = pearsonr(la_rents_initial, la_rents_lag)

# Check if the p-value is significant at the 5% level
print(p_value<0.05)
print(r)
```

Autocorrelation is a powerful tool as it helps you understand to what the extent the past can explain the future. Notice how there is a very strong relationship between the rent prices one year in the past versus the present. This tells you that rents year-over-year tend to follow very similar patterns.

Since the data is autocorrelated, the assumption of independence in a t-test would be violated.
